# Sensors, Locations and Maps

## Sensors

Sensors are hardware devices that measure the physical environment.

- Motion - 3-axis Accelerometer
- Position - 3-axis Magnetic field
- Environment - Pressure

### ￼￼￼￼SensorManager

System service that manages sensors.
Get instance with

- getSystemService(Context.SENSOR_SERVICE )

Access a specific sensor with

- SensorManager.getDefaultSensor(int type)

Sensor type constants are:

- Accelerometer - Sensor.TYPE_ACCELEROMETER
- Magnetic field - Sensor.TYPE_MAGNETIC_FIELD
- Pressure - Sensor.TYPE_PRESSURE

### ￼SensorEventListener

Interface for SensorEvent callbacks.

Called when the accuracy of a sensor has changed

- void onAccuracyChanged(Sensor sensor, int accuracy)

Called when sensor values have changed

- void onSensorChanged(SensorEvent event)

Use the SensorManager to register/ unregister for SensorEvents.

To register a SensorEventListener for a given sensor

- public boolean registerListener (SensorEventListener listener,Sensor sensor, int rate)

Unregisters a listener for the sensors with which it is registered

- public void unregisterListener (SensorEventListener listener,Sensor sensor)

Ideal values should be X = 0, Y = 9.81 and Z = 0

But these values will vary due to natural movements, non-flat surfaces, noise, etc.

But we can fix this by filtering the accelerometer values

There are two common types of filters:

- Low-pass filter (Deemphasize transient force changes Emphasize constant force components)
- High-pass filter (Emphasize transient force changes Deemphasize constant force components)

Next we can examine the **SensorFilteredAccelerometer** application

Apply both a low-pass and a high-pass filter to raw accelerometer values and displays the filtered values
