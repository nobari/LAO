There are two types of Intents:
- Explicit activation (if you know which activity to start)
- Implicit activation (if you know the action which you want to perform, but give the user an ability to choose an app to perform that action)

So, for explicit activation the user will immediately see the next activity (because the developed specified it explicitly in the code). In the case of implicit activation the user will have the choice which will look like the following:

```python
Image(filename='screen1.png')
```

Intents contain additional fields with parameters which you can specify. For example, you can spesify Intent which redirects the user to the browser and the website address would be one of the parameters. Or you can redirect him to a map and transfer location as an Intent parameter.

The whole list of Intent fields/parameters:
- Action - desired operation. E.g. ACTION_DIAL to dial a number
- Data - some data, e.g. map location or a phone number
- Category - additional information about components which can handle the intent. E.g. CATEGORY_BROWSABLE
- Type - type of the content. E.g. "text/plain"
- Component - who should receive the intent (context and class)
- Extras - a map with additional information. E.g. Email title
- Flags - how the intent should be handled. E.g. FLAG_ACTIVITY_NO_HISTORY

- onCreate()
- onCreateView()
- onActivityCreated()
- onStart()
- onResume()
- onPause()
- onStop()
- onDestroy()
- onDestroyView()
- onDetach()


How to create a fragment?
There are two ways:
- static (XML layout). You simply declare fragment in a separate XML layout file.
- programmatic (using code). As usual and as it happens with other UI elements, you can create fragments manually during the runtime. You will need to create an object of Fragment class, add some UI elements to that class and finally to attach the fragment to a certain Activity.
