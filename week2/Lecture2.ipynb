### Step 1. Defining Resources

Non-source code entities, that allow to customize for different devices and users. Layout, Strings, Images, Menues, Animations.

### Strings

Types: strings, string arrays, plurals. Typically stored in res/values/*.xml. E.g. Language and layout. Accessed by: @string/string_name. See example in MapLocation with Italian language.

### User Interface Layout

Stored in res/layout/*.xml. Accessed in Java as R.layout.layout_name. See example in MapLocation.

### R.java

At compilation time, resources are used to create R.java class. Java code uses R.class to access resources.

Navigation through activities in Android is done via Tasks, the Task Backstack, suspending and resuming activities.

### Tasks

Task is a set of related activities. Can span multiple applications. Start at home screen. Touch an app - new task started. Touch HOME button - task is closed.

### Task Back Stack

When activity launches, it is pushed to the top of the backstack for current task. When destroyed - it is popped off the backstack. Push/pop style.

### onCreate()

 - called when activity is created
 - sets up an initial state
 - calls 4 functions: super.onCreate() - Android does it's own initialization, set conetnt view (what GU should be), retain references to UI views and elements, configure them as required.

```python
Image(filename='oncreate.png') 
```

### onRestart()

Special preprocessing needed only when being restarted.

### onStart()

Special preprocessing needed only when becoming visible, starting  only-visible behavior (sensors, get data)

### onResume()

Foreground only behavior - e.g. starting animations.

### onPause()

Shut down foreground only behaviors, when is about to focus to another activity. Save current state.

### onStop()

No longer visible, caches state. Can be restarted. !!! If Android kills your app due to low memory, it cannot be restarted, because this method may not be called.

### onDestroy()

Activity is about to be destroyed, Activity resources are released (private threads are shut down). Same note as before.
