## Common view operations

* Set visibility 
   -  Show
   -  Hide
* Set checked state
* Set listeners
   -  Executed when event
* Set properties
   -  Opacity
   -  Background
   -  Rotation
* Input focus


#### View event sources

* User
   -  Touch
* System
   -  Keyboard



### Handling events

* Attach listeners to events
* Numerous listeners for view which depend on action
   -  Click
   -  Long click
   -  On Focus
   -  On Key


### Displaying view

* View organized in a tree
* Displaying has steps
   -  Measure (get dimensions)
   -  Layout (position)
   -  Draw


### Handling new events

Custom view can override
* onMeasure()
* onLayout()
* onDraw()
* onFocusChanged()
* onKeyUp()
* onKeyDown()
* onWindowVisibilityChange()

### ViewGroup

* Invisible view which contains other views
* Used for grouping and organizing
* Base class for view containers and layouts
